### Dealing with missing values

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.isna().transpose(),
            cmap="crest",
            cbar_kws={'label': 'Missing Data'})
plt.savefig("visualizing_missing_data_with_heatmap_Seaborn_Python.png", dpi=100)
```

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

```python
df.drop('id',inplace=True,axis=1)
```

```python
df.head()
```

```python
df.isna().sum()
```

### Select the Features and Label

```python
sns.countplot(x = 'diagnosis', data = df)
# x = data.iloc[:, 3:]
y = df.diagnosis
y = y.map({'M':1,'B':0})
```

```python
x=df.drop('diagnosis',axis=1)
```

```python
x
```

### Check the Data type of the Data set

```python
x.dtypes
```

### Split the data set 

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=365)
```

### import GradientBoostingClassifier

```python
from sklearn.ensemble import GradientBoostingClassifier
```

### Train the Model

```python
gradient_booster = GradientBoostingClassifier(learning_rate=0.1)
```

```python
gradient_booster.fit(x_train,y_train)
```

```python
#Calculating Score
gradient_booster.score(x_test,y_test)
```

```python
yhat=gradient_booster.predict(x_test)
```

### Classfication Report

```python
from sklearn.metrics import classification_report

# target_names = ["benig","malignant"]
print(classification_report(y_test, yhat))
```

```python
from sklearn.metrics import confusion_matrix
# y_pred = model.predict(X_test)
cm = confusion_matrix(y_test, yhat)
cm
```

### Confusion matrix

```python
plt.figure(figsize=(7,5))
sns.heatmap(cm, annot=True,fmt='d',cbar=False,cmap='BuPu')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
